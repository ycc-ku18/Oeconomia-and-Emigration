```python
#reading the lattitude and longitude of the countries and the amount of remittance
df = pd.read_csv('latlon.csv')
df =df.fillna(0)
df
```

```python
# Map of Nepal
m = Basemap(projection='mill',
            llcrnrlat = 25,
            llcrnrlon = 79,
            urcrnrlat = 32,
            urcrnrlon = 90,
           resolution='h')
m.drawcoastlines()
m.drawcountries(linewidth=2)
m.bluemarble()

plt.title('Our Country')
plt.show()
```

```python
#Map of World with Nepal
m = Basemap(projection='mill',
            llcrnrlat = -90,
            llcrnrlon = -180,
            urcrnrlat = 90,
            urcrnrlon = 180,
           resolution='l')
m.drawcoastlines()
m.drawcountries(linewidth=0.5)
m.fillcontinents(color='#29832C', lake_color='#33BBFF')
m.drawmapboundary(fill_color='#33BBFF')
nepallon=84.1
nepallat=28.
xpt,ypt = m(nepallon,nepallat)
m.plot(xpt,ypt,'ro', markersize=5)
plt.show()
```

```python
#countries the remmitance comes from.
m.fillcontinents(color='#29832C', lake_color='#33BBFF')
m.drawmapboundary(fill_color='#33BBFF')
m.drawcountries(linewidth=0.5)
m.drawcoastlines()
latti=df["Latitude"]
longi=df["Longitude"]
invalues=df["receiving"]
for x in range(0 , 33):
    if(invalues[x]!=0):
        
        lon= longi[x]
        lat= latti[x]
        xpt,ypt = m(lon,lat)
        m.plot(xpt,ypt,'yo', markersize=6)
        
plt.show()
```

```python
# Countries the remittance goes to.
m.fillcontinents(color='#29832C', lake_color='#33BBFF')
m.drawmapboundary(fill_color='#33BBFF')
m.drawcountries(linewidth=0.5)
m.drawcoastlines()
latti=df["Latitude"]
longi=df["Longitude"]
outvalues=df["sending"]
for x in range(0 , 33):
    if(outvalues[x]!=0):
        lon= longi[x]
        lat= latti[x]
        xpt,ypt = m(lon,lat)
        m.plot(xpt,ypt,'yo', markersize=8)
        m.drawgreatcircle(nepallon, nepallat,lon,lat, color='y')
plt.show()
```

```python
#creating dataframe
df = pd.read_csv("Number of Nepalese going aboard by purpose of visit and month, 2012-2013.csv")
```

```python
#function to calculate the purpose as it is divided into month
def total_calc(my_df,my_label,my_index):
    dflabel=my_df.loc[my_label]
    total = dflabel[my_index].sum()
    return total   
    
```

```python
df.iloc[:,[0,4]]
```

```python
df_both = df.iloc[0:,[0,4]]
```

```python
df_both.set_index("Purpose of visit", inplace = True)
```

```python
df_both.index.unique()
```

```python
a = "Nepalese going aboard by purpose of visit and month"
```

```python
entertainmet_total_both = total_calc(df_both,"Entertainment",a)
health_total_both = total_calc(df_both,"Health Check up",a)
business_total_both = total_calc(df_both,"Business",a)
pilgrimage_total_both = total_calc(df_both,"Pilgrimage",a)
official_total_both = total_calc(df_both,"Official",a)
conference_total_both = total_calc(df_both,"Conference",a)
study_total_both = total_calc(df_both,"Study/Training",a)
employment_total_both = total_calc(df_both,"Employment",a)
sports_total_both = total_calc(df_both,"Sports",a)
others_total_both = total_calc(df_both,"Others",a)
notspecified_total_both = total_calc(df_both,"Not Specified",a)
foreign_emp_total_both = total_calc(df_both,"Foreign Employment",a)

total_arr = [entertainmet_total_both,health_total_both,business_total_both,conference_total_both,study_total_both,employment_total_both+foreign_emp_total_both,pilgrimage_total_both,others_total_both,notspecified_total_both,sports_total_both,official_total_both]
```
